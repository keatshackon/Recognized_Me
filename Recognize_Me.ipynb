```python
# from keras.datasets import mnist,cifar10
import matplotlib.pyplot as plt
import cv2
# from PIL import Image
import numpy as np
from keras.utils import np_utils
# example of converting an image with the Keras API
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import array_to_img
```

```python
train_ld = []
test_ld = []
for i in range(1,203):
    if(i <= 132):
        train_ld.append(1)
    else:
        train_ld.append(0)

for i in range(1,31):
    if(i <= 12):
        test_ld.append(1)
    else:
        test_ld.append(0)
```

```python
print("Integer based index : ")
print(train_ld[0:10])
print(train_ld[150:170])

train_ld = np_utils.to_categorical(train_ld,2)

print("One hot Encoding : ")
print(train_ld[0:5]) 
print(train_ld[150:160])
```

```python
print("Integer based index : ")
print(test_ld[0:13])

test_ld = np_utils.to_categorical(test_ld,2)

print("One hot Encoding : ")
print(test_ld[0:13])
print(test_ld[13:31])
```

```python
train_d = [img_to_array(cv2.imread("images/train/" + str(i) + ".jpg"))  for i in range(1,203)]
test_d = [img_to_array(cv2.imread("images/test/" + str(i) + ".jpg"))  for i in range(1,31)]
```

```python
len(train_d)
train_d = np.asarray(train_d)
test_d = np.asarray(test_d)
print(train_d.shape)
print(test_d.shape)

train_d = train_d.astype('float32')/255
test_d = test_d.astype('float32')/255
```

```python
# print(train_d[0])
# print(test_d[0])
```

```python
plt.imshow(train_d[120])
plt.show()
```

```python
fig = plt.figure(figsize=(20,20))
cnt = 125
for i in range(30):
    ax = fig.add_subplot(6, 5, i+1, xticks=[], yticks=[])
    ax.imshow(train_d[cnt], cmap='gray')
    
    ll = train_ld[cnt]
    if(ll[1] == 1):
        ax.set_title("Recognized Mr.Keats")
    else:
        ax.set_title("Didn't Recognized Mr.Keats")
    cnt += 1
        
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten

# define the model
model = Sequential()
model.add(Flatten(input_shape= train_d.shape[1:]))
model.add(Dense(70, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(30, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(2, activation='softmax'))

# summarize the model
model.summary()
```

```python
# compile the model
model.compile(loss='categorical_crossentropy', optimizer='rmsprop', 
              metrics=['accuracy'])
```

```python
# evaluating test accuracy before Training
score = model.evaluate(test_d,test_ld, verbose=0)
accuracy = 100 * score[1]

# print test accuracy
print('Test accuracy: %.4f%%' % accuracy)
```

```python
from keras.callbacks import ModelCheckpoint   

# train the model
checkpointer = ModelCheckpoint(filepath='keats.model.best.hdf5', 
                               verbose=1, save_best_only=True)
hist = model.fit(train_d,train_ld, batch_size=30, epochs=6,
          validation_split=0.1, callbacks=[checkpointer],
          verbose=1, shuffle=True)
```

```python
model.load_weights('keats.model.best.hdf5')
```

```python
# evaluate test accuracy
score = model.evaluate(test_d, test_ld, verbose = 0)
accuracy = 100*score[1]

# print test accuracy
print('Test accuracy: %.4f%%' % accuracy)
```

```python
# get predictions on the test set
y_hat = model.predict(test_d)
# print(y_hat[:31])

predict_l = ['Not Mr.Keats', 'Recognize Mr.keats']
```

```python
fig = plt.figure(figsize=(20, 20))
for i in range(0,30):
    ax = fig.add_subplot(6, 5, i + 1, xticks=[], yticks=[])
    ax.imshow(np.squeeze(test_d[i]))
    pred_idx = np.argmax(y_hat[i])
    true_idx = np.argmax(test_ld[i])
    ax.set_title("{} ({})".format(predict_l[pred_idx],predict_l[true_idx]),
                 color=("green" if pred_idx == true_idx else "red"))
    
```
